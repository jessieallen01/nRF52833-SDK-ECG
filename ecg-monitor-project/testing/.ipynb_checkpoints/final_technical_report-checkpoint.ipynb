# Part 1: Battery Voltage Measurement

### Introduction/Methods

To test the accuracy of the battery level measurement across the input voltage range of 0-3V, we implemented a linear mapping in the code that converts this voltage range to a 0-100% duty cycle, generating the linear equation $y=\frac{x}{3}$. Our testing approach utilizes a DC power supply connected to the AIN0 channel of the nRF board, with voltage levels changed to observe the resulting duty cycle output. To ensure data accuracy and precision, we applied three independent measurement methods: logging statements, Bluetooth outputs, and direct oscilloscope readings to determine % duty cycle values.

From the collected data, we calculated basic statistical measures including mean and standard deviation, and determined a 95% confidence interval to assess measurement precision. We then generated linear regression models for each of the three measurement methods to quantitatively compare how closely the regression lines align with the linear mapping implemented in the code.

# Part 2:  Taking Heart Rate Measurements 
### Introduction/Methods

This lab section implements a heart rate monitor that reads an ECG signal via differential asynchronous ADC, processes it in real time, and modulates an LED's blinking frequency to simulate a heart rate display at a 25% duty cycle across a 40180 bpm range.

Several methods were considered, including a two buffer system, ring buffers, and timer-based approaches. However, a single-buffer technique proved most streamlined in this case. The buffer operates as follows: once filled, it triggers a buffer full event to begin measurements. As data collection continues, new ADC values overwrite the buffer via a callback function, which enables the "real-time signal processing" aspect without excessive overhead.

A timer is initiated when the user activates the heart rate monitoring state via a button press. At specified intervals defined by a macro (4 seconds in this case), the on and off times are processed and sent to a thread to modulate the LED blinking rate. The heart rate is determined by counting the number of cycles within the time window. This is calculated each time the signal crosses the threshold when the ECG signal R wave is ascending, resulting in the cycle counter incrementing by 1.

The ECG signal has a peak-to-peak voltage of 1V, with the positive peak measured around  500mV. The threshold (300 mV in our case) was positioned to minimize noise interference by avoiding the signal extrema (peaks and troughs) as well as the baseline crossing region (where the P, Q, S, T, and U waves reside). This placement ensures reliable cycle detection without triggering false counts.

In terms of testing methods, a wave form generator was used to produce an ECG signal in the AIN1 and AIN2 channels at 40, 60, 120, 150, and 180 bpm (4 samples per heart rate level). The same three methods (logging statements, bluetooth, and oscilloscope) used in the battery measurement were also employed to determine the accuracy and precision of our implemented system. Basic stats were taken in addition to a 95% confidence interval to display the precision of the data while a linear regression was calculated to identify the accuracy. Note that the 25% duty cycle was also analyzed in a similar manner for precision and accuracy for both the logging statements and oscilloscope output.

## Heart Rate Accuracy Conclusions

After evaluating methods for continuously updating the LED frequency as the ECG signal is acquired, a single-buffer timer approach was determined to be optimal for this application.

Statistical testing with 95% confidence intervals and basic statistics showed that all three measurement methods, logging statements, Bluetooth, and oscilloscope output, yielded identical values for both the heart rate and 25% duty cycle. This consistency demonstrates that the precision of the heart rate calculation is high across all tested methods.

Linear regression analysis comparing actual to experimental values showed strong accuracy across most of the tested range. However, the 40 bpm measurement deviated from the expected value, indicating a systematic error at lower heart rates. This discrepancy stems from a fundamental limitation: lower heart rates produce fewer beats within the sampling window, requiring a larger window to capture sufficient cycles. Conversely, extending the sampling window compromises the real-time updating capability, as values refresh less frequently.

### Root Cause Analysis
A 40 bpm signal completes one cycle every 1.5 seconds, so a 4-second sampling window should  capture approximately 2.67 cycles. However, logging statements consistently reported 3 cycles during this interval, yielding a calculated rate of 45 bpm. The most likely cause is the cycle detection mechanism: since cycles are counted on each ascending R-wave threshold crossing, a partial wave (the remaining 0.67 of the third cycle) is counted as a complete cycle, introducing the observed error. This issue can be mitigated by implementing a dual-threshold condition that only increments the cycle counter when both the ascending and descending portions of the R wave cross the threshold. 

Overall, the implemented system demonstrates high accuracy and precision across the 40180 bpm range, with the exception of the 40 bpm measurement can be mended by changing the threshold conditions or making slight adjustments to the sampling window.
