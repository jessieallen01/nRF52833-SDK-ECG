## Results

Loading in .csv files for both logging and oscilloscope data

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats


# Load the data

duty_data = pd.read_csv('kernels_duty_hbeat_cycle.csv')

# Print the first few rows of the data
duty_data.head(20)
```

## Analyzing both ON and OFF cycles
Note that we are analyzing the data by the AFTER column to get a better idea of the general statistics results:

```python
duty_data['Duty Cycle OFF After (ms)'].describe().to_frame().T
```

```python
duty_data['Duty Cycle ON After (ms)'].describe().to_frame().T
```

### How to Calculate 95% Confidence Interval for a Mean
Given the variation in the data, confidence intervals were calculated for both the on and off duty cycles.

```python
x = duty_data['Duty Cycle OFF After (ms)']


mean = np.mean(x)
sem = stats.sem(x)  # Standard error of the mean
confidence_interval = stats.t.interval(0.95, len(x) - 1, loc=mean, scale=sem)

print("Mean:", mean)
print("95% confidence interval:", confidence_interval)
```

```python
x = duty_data['Duty Cycle ON After (ms)']


mean = np.mean(x)
sem = stats.sem(x)  # Standard error of the mean
confidence_interval = stats.t.interval(0.95, len(x) - 1, loc=mean, scale=sem)

print("Mean:", mean)
print("95% confidence interval:", confidence_interval)
```

Note the confidence intervals are quite small (this is good news!). Now we have to note if there is any change between the 'before' and 'after' data.

```python
columns_to_plot_OFF = [
    'Duty Cycle OFF Before (ms)', 
    'Duty Cycle OFF After (ms)'
]

duty_data[columns_to_plot_OFF].plot(kind='box', figsize=(8, 5))
plt.title('Boxplot of OFF Duty Cycle Columns')
plt.ylabel('Milliseconds')
plt.xticks(rotation=45)
plt.show()
```

Although the data is spread out for the 'after cycle' the distribution is only .3 milliseconds. When noting the difference between the averages of both groups, we can see it is less than .05 milliseconds which is almost negligible.

```python
columns_to_plot_ON = [
    'Duty Cycle ON Before (ms)',
    'Duty Cycle ON After (ms)'
]
duty_data[columns_to_plot_ON].plot(kind='box', figsize=(8, 5))
plt.title('Boxplot of ON Duty Cycle Columns')
plt.ylabel('Milliseconds')
plt.xticks(rotation=45)
plt.show()
```

Again, the spread of both groups is slightly varied but only by .3 milliseconds. The averages, however, are almost identical.
